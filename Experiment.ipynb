```python
import matplotlib.pyplot as plt
import pickle
import numpy as np

def hypercube_plot(path="pickle/hypercube.pickle"):
    pickle_in = open(path, "rb")
    fig, ax = plt.subplots(figsize=(5, 5))
    k_star = pickle.load(pickle_in)
    k_star_max = pickle.load(pickle_in)
    d = pickle.load(pickle_in)
    x = np.arange(1, d + 1)
    
    for k in range(k_star, k_star_max):
        y = pickle.load(pickle_in)
        std = pickle.load(pickle_in)
        color = ((k - k_star) / ((k_star_max - 1) - (k_star)),
                 0,
                 1 - (k - k_star) / ((k_star_max - 1) - (k_star)),
                 1)

        if ((k - k_star) % 2) == 0:
            ax.plot(x, y,
                    label="k = " + str(k),
                    color=color)
            ax.fill_between(x, y - std,
                            y + std,
                            alpha=0.2, edgecolor=color,
                            facecolor=color,
                            antialiased=True)
    
    M_rank = pickle.load(pickle_in)
    
    ax.set_xlabel('Eigenvalues', size=12)
    ax.set_ylabel('Magnitude of the Eigenvalues', size=12)
    ax.legend(loc=0)
    plt.savefig("hypercube/Recover_M_eigen_diff_k.pdf")
    plt.show()
    
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.set_xlabel('Dimension k ', size=20)
    ax.set_ylabel('Rank of M', size=20)
    ax.plot(list(np.arange(k_star, k_star_max)), M_rank)
    plt.savefig("hypercube/Recover_M_dim.pdf")
    plt.show()
    
    # ALL THE RAW DATA ARE IN M BELLOW FOR POTENTIAL OTHER PLOT
    M = pickle.load(pickle_in)
    
    pickle_in.close()
```

```python
hypercube_plot()
```

```python
import matplotlib.pyplot as plt
import pickle
import numpy as np

def plot_graph(path="pickle/sunset.pickle", path_save="color_transfer/sunset.pdf"):
    
    pickle_in = open(path, "rb")
    diff_W = pickle.load(pickle_in)
    loop_order = pickle.load(pickle_in)
    pickle_in.close()
    
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.set_xticks([0, 49, 99])
    ax.set_yticks([diff_W.min(), diff_W.mean(), diff_W.max()])
    ax.set_xticklabels([1, 50, 100], size=12)
    ax.set_yticklabels(np.round([diff_W.min(), diff_W.mean(), diff_W.max()], 4), size=12)

    plt.step(np.arange(len(diff_W)), diff_W[loop_order], "g")
    plt.plot([0], (diff_W[loop_order])[0], "o", markersize=8, color=(1, 0, 0),
             label="Most stable matrix")
    
    plt.plot([(len(diff_W) - 1)//2], (diff_W[loop_order])[(len(diff_W) - 1)//2], "o",
            markersize=8, color=(0.5, 0, 0.5), label="Least stable mahalanobis matrix")
    
    plt.plot([len(diff_W) - 1], (diff_W[loop_order])[len(diff_W) - 1], "o",
            markersize=8, color=(0, 0, 1), label="Least stable matrix")

    plt.plot(np.where(loop_order==0), diff_W[0], "s",
            markersize=8, color=(0, 0, 0), label="Euclidean distance cost matrix")
    
    ax.legend(fontsize=12)
    leg = ax.get_legend()
    ax.set_ylabel('Wasserstein stability', size=20)
    ax.set_xlabel('Various cost matrices', size=20)
    plt.setp(ax.get_yticklabels(), rotation=45, ha="right", rotation_mode="anchor")
    
    plt.savefig(path_save, bbox_inches='tight')
    
    plt.show()
```

```python
plot_graph(path="pickle/sunset.pickle", path_save="color_transfer/sunset.pdf")
```

```python
import matplotlib.pyplot as plt
import pickle
import numpy as np

def plot_correlation(path="pickle/mnist.pickle", path_save="gaussian_mnist/noise_mnist.pdf"):
    pickle_in = open(path, "rb")
    W_stable = pickle.load(pickle_in) # Difference of Wasserstein related to noise 
    worst_W_ = pickle.load(pickle_in) # Difference of Wasserstein related to stability in the initial order
    worst_W_stable = pickle.load(pickle_in) # Difference of Wasserstein related to stability
    stable = pickle.load(pickle_in) # stable is used earlier only to avoid crash.
    order_plot = pickle.load(pickle_in) # The original order of the matrices. Used to find the position
    # of the euclidean based matrix
    W_ = pickle.load(pickle_in) # Difference of Wasserstein related to noise in the initial order
    pickle_in.close()
    
    fig, ax1 = plt.subplots(figsize=(10, 5))
    ax1.set_xlabel('Various cost matrices', size=20)
    ax1.ticklabel_format(axis='y', style='sci')
    color = (1, 0, 0)
    ax1.set_ylabel('Noise stability', color=color, size=20)
    ax1.plot([i for i in range(np.sum(stable))], W_stable, "+", color=(0, 0, 0),
             label="Various cost matrices")
    ax1.plot([i for i in range(np.sum(stable))], W_stable, "+", color=color)
    ax1.plot(np.where(order_plot == 50)[0][0], W_[-1], "o", color=(0, 0, 0), markersize=8,
             label="Euclidean distance cost matrix")
    ax1.plot(np.where(order_plot == 50)[0][0], W_[-1], "o", color=color, markersize=8)
    ax1.legend(loc=0, fontsize=15)
    ax1.tick_params(axis='y', labelcolor=color)
    plt.setp(ax1.get_yticklabels(), rotation=45, ha="right", rotation_mode="anchor")

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
    ax2.ticklabel_format(axis='y', style='sci')
    color = (0, 0, 1)
    ax2.set_ylabel('Wasserstein stability', color=color, size=20)
    plt.plot([i for i in range(np.sum(stable))], worst_W_stable, "+", color=color)
    plt.plot(np.where(order_plot == 50)[0][0], worst_W_[-1], "o", color=color, markersize=8)
    ax2.tick_params(axis='y', labelcolor=color)
    plt.setp(ax2.get_yticklabels(), rotation=-45, ha="left", rotation_mode="anchor")
    
    fig.tight_layout()

    plt.savefig(path_save, bbox_inches="tight")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def plot_correlation_paper(res=["pickle/gaussian.pickle", "pickle/mnist.pickle"]):
    

    Wst = []
    Ww = []
    Wws = []
    S = []
    Op = []
    W = []

    scaler = MinMaxScaler()

    for k in res:
        pickle_in = open(k, "rb")
        W_stable = pickle.load(pickle_in)
        worst_W_ = pickle.load(pickle_in)
        worst_W_stable = pickle.load(pickle_in)
        stable = pickle.load(pickle_in)
        order_plot = pickle.load(pickle_in)
        W_ = pickle.load(pickle_in)
        pickle_in.close()

        Wst.append(scaler.fit_transform(W_stable.reshape(-1, 1)))
        Ww.append(scaler.fit_transform(worst_W_.reshape(-1, 1)))
        Wws.append(scaler.fit_transform(worst_W_stable.reshape(-1, 1)))
        S.append(scaler.fit_transform(stable.reshape(-1, 1)))
        Op.append(scaler.fit_transform(order_plot.reshape(-1, 1)))
        W.append(scaler.fit_transform(W_.reshape(-1, 1)))



    fig, ax = plt.subplots(figsize=(7, 5))

    ax.set_xlabel('Noise sensitivity', size=20, labelpad=15)
    ax.set_ylabel('Wasserstein (in)stability', size=20) 
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)

    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    plt.xticks([])
    plt.yticks([])

    cax = ax.scatter(Wws[0], Wst[0]+0.5, c = Wst[0], cmap = 'bwr', marker = 'o', s = 60, label = 'Gaussians data set')

    plt.scatter(Wws[1], Wst[1], c = Wst[1], cmap = 'bwr', marker = 'd', s = 60, label = 'MNIST data set')
    car = plt.colorbar()
    car.set_ticks([np.min(Wst[1]),np.max(Wst[1])])
    car.set_ticklabels(['low','high'])
    car.ax.tick_params(labelsize=15)

    leg = plt.legend(loc="upper left", fontsize=15)
    for marker in leg.legendHandles:
        marker.set_color('k')

    plt.savefig("gaussian_mnist/gaussians+mnist.pdf", dpi = 600, bbox_inches='tight')
    plt.show()
```

```python
plot_correlation_paper()
```
